```python
import re
from typing import List, Dict, Set

class SpecializationGrouper:
    def __init__(self):
        self.specialization_groups = {
            # Business & Management
            'business': re.compile(r'\b(business|commercial|management|enterprise|affaires|développement|stratég)', re.IGNORECASE),
            
            # Marketing & Communication
            'marketing': re.compile(r'\b(marketing|communication|brand|marque|publicitaire|promotion|commercial)', re.IGNORECASE),
            
            # Finance & Banking
            'finance': re.compile(r'\b(finance|banking|banque|audit|comptabil|contrôle|gestion|investment|patrimoine|fiscale)', re.IGNORECASE),
            
            # Digital & Technology
            'digital': re.compile(r'\b(digital|numérique|cyber|informatique|tech|IT|web|cloud|data|intelligence)', re.IGNORECASE),
            
            # Data & AI
            'data_ai': re.compile(r'\b(data|intelligence|artificial|ia|analytics|big\s*data|machine\s*learning|ai)', re.IGNORECASE),
            
            # Engineering & Technical
            'engineering': re.compile(r'\b(engineer|ingénieur|génie|civil|mechanical|mécanique|robotique|robo|automation)', re.IGNORECASE),
            
            # Supply Chain & Logistics
            'supply_chain': re.compile(r'\b(supply|chain|logist|achat|procurement|purchasing|transport)', re.IGNORECASE),
            
            # Human Resources
            'human_resources': re.compile(r'\b(ressources\s*humaines|human\s*resources|rh|hr|talent|people)', re.IGNORECASE),
            
            # Luxury & Fashion
            'luxury': re.compile(r'\b(luxury|luxe|fashion|mode|premium|prestige|craft)', re.IGNORECASE),
            
            # International Business
            'international': re.compile(r'\b(international|global|export|import|geopolitics|intercultural)', re.IGNORECASE),
            
            # Entrepreneurship
            'entrepreneurship': re.compile(r'\b(entrepreneur|startup|innovation|création|développement\s*des\s*affaires)', re.IGNORECASE),
            
            # Hospitality & Tourism
            'hospitality': re.compile(r'\b(hospitality|tourism|tourisme|hotel|hôtel|restaurant|gastronomie|wine|vin)', re.IGNORECASE),
            
            # Sports & Entertainment
            'sports': re.compile(r'\b(sport|entertainment|événement|event|spectacle|culture)', re.IGNORECASE),
            
            # Real Estate
            'real_estate': re.compile(r'\b(immobilier|real\s*estate|property|patrimoine|construction|bâtiment)', re.IGNORECASE),
            
            # Health & Medicine
            'health': re.compile(r'\b(health|santé|medical|médical|biotech|pharma|nutrition)', re.IGNORECASE),
            
            # Design & Creative
            'design': re.compile(r'\b(design|créat|art|graph|illustration|visual|motion|animation|3d|2d)', re.IGNORECASE),
            
            # Media & Journalism
            'media': re.compile(r'\b(media|médias|journalism|journalisme|communication|audiovisuel|vidéo)', re.IGNORECASE),
            
            # Environment & Sustainability
            'sustainability': re.compile(r'\b(sustainable|durable|environment|écologique|green|transition|rse)', re.IGNORECASE),
            
            # Cybersecurity
            'cybersecurity': re.compile(r'\b(cybersécurité|cybersecurity|sécurité|security|protection)', re.IGNORECASE),
            
            # Energy & Environment
            'energy': re.compile(r'\b(energy|énergie|nuclear|nucléaire|renewable|smart\s*grid|power)', re.IGNORECASE),
            
            # Gaming & Interactive
            'gaming': re.compile(r'\b(game|gaming|interactive|immersive|virtual|reality|vr|ar)', re.IGNORECASE),
            
            # Food & Agriculture
            'food': re.compile(r'\b(food|agriculture|agri|culinary|culinaire|pâtisserie|chocolat)', re.IGNORECASE),
            
            # Education & Training
            'education': re.compile(r'\b(education|formation|training|pédagogie|enseignement)', re.IGNORECASE),
            
            # Legal & Compliance
            'legal': re.compile(r'\b(legal|juridique|droit|law|compliance|conformité|risk|risque)', re.IGNORECASE),
            
            # Aviation & Transportation
            'aviation': re.compile(r'\b(aviation|aéronautique|aerospace|transport|mobility|mobilité)', re.IGNORECASE),
            
            # Architecture & Construction
            'architecture': re.compile(r'\b(architecture|construction|building|urbanisme|aménagement)', re.IGNORECASE),
            
            # Chemistry & Materials
            'chemistry': re.compile(r'\b(chemistry|chimie|materials|matériaux|biotechnology|biotech)', re.IGNORECASE)
        }
    
    def categorize_specialization(self, specialization: str) -> List[str]:
        """
        Categorize a single specialization string into one or more categories.
        
        Args:
            specialization (str): The specialization string to categorize
            
        Returns:
            List[str]: List of matching categories
        """
        categories = []
        
        for category, regex_pattern in self.specialization_groups.items():
            if regex_pattern.search(specialization):
                categories.append(category)
        
        return categories if categories else ['other']
    
    def group_specializations(self, specializations: List[str]) -> Dict[str, List[str]]:
        """
        Group a list of specializations by category.
        
        Args:
            specializations (List[str]): List of specialization strings
            
        Returns:
            Dict[str, List[str]]: Dictionary with categories as keys and lists of specializations as values
        """
        grouped = {}
        
        for spec in specializations:
            categories = self.categorize_specialization(spec)
            for category in categories:
                if category not in grouped:
                    grouped[category] = []
                grouped[category].append(spec)
        
        return grouped
    
    def get_unique_categories(self, specializations: List[str]) -> Set[str]:
        """
        Get all unique categories found in the specializations list.
        
        Args:
            specializations (List[str]): List of specialization strings
            
        Returns:
            Set[str]: Set of unique categories
        """
        categories = set()
        
        for spec in specializations:
            spec_categories = self.categorize_specialization(spec)
            categories.update(spec_categories)
        
        return categories
    
    def filter_by_category(self, specializations: List[str], category: str) -> List[str]:
        """
        Filter specializations by a specific category.
        
        Args:
            specializations (List[str]): List of specialization strings
            category (str): Category to filter by
            
        Returns:
            List[str]: List of specializations matching the category
        """
        if category not in self.specialization_groups:
            return []
        
        filtered = []
        regex_pattern = self.specialization_groups[category]
        
        for spec in specializations:
            if regex_pattern.search(spec):
                filtered.append(spec)
        
        return filtered

# Example usage
if __name__ == "__main__":
    # Create an instance of the grouper
    grouper = SpecializationGrouper()
    
    # Example specializations from your list
    sample_specializations = [
        "robotics"
    ]
    
    # Group specializations
    grouped = grouper.group_specializations(sample_specializations)
    
    # Print results
    for category, specs in grouped.items():
        print(f"\n{category.upper()}:")
        for spec in specs:
            print(f"  - {spec}")
    
    # Example of categorizing a single specialization
    single_spec = "Digital Marketing & Business"
    categories = grouper.categorize_specialization(single_spec)
    print(f"\nCategories for '{single_spec}': {categories}")
    
    # Get unique categories
    unique_cats = grouper.get_unique_categories(sample_specializations)
    print(f"\nUnique categories found: {sorted(unique_cats)}")
```
